```python
from selenium import webdriver
import time
import chromedriver_binary
import numpy as np
import pandas as pd
from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
# Example for Chrome
driver = webdriver.Chrome()
```

```python
driver.get('https://www.youtube.com/c/GeeksforGeeksVideos/videos')
```

```python
soup = BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup.prettify()
```

```python
sp = soup.find_all('a',class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media')    
```

```python
title=[]
link=[]
views=[]

for i in sp:
    try:
        title.append(i['title'])
    except:
        title.append(np.nan)
    try:
        link.append("https://www.youtube.com"+i['href'])
    except:
        link.append(np.nan)
sp1=soup.find_all('span',class_='inline-metadata-item style-scope ytd-video-meta-block')
for i in sp1:
    try:
        views.append(i.text)
    except:
        views.append(np.nan)

print(len(title))
print(len(link))
print(views)
view=[]
dayupload=[]
for i in range(56):
    if(i%2==0):
        view.append(views[i])
    else:
        dayupload.append(views[i])
print(len(view))
print(len(dayupload))
```

```python
print(title)
print(link)
print(view)
print(dayupload)
```

```python
data=[]
for i in range(28):
    exp=[]
    exp.append(title[i])
    exp.append(link[i])
    exp.append(view[i])
    exp.append(dayupload[i])
    data.append(exp)
    
df=pd.DataFrame(data,columns=['title','link','view','dayupload'])
# data_csv=df.to_csv(data,index=False)
```
